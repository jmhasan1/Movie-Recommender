### A helper function is a function that performs part of the computation of another function. Helper functions are used to make your programs easier to read by giving descriptive names to computations.

### The ast module (Abstract Syntax Tree) allows us to interact with and modify Python code

### The ast.literal_eval method is one of the helper functions that helps traverse an abstract syntax tree. This function evaluates an expression node or a string consisting of a Python literal or container display.

### The ast.literal_eval method can safely evaluate strings containing Python values from unknown sources without us having to parse the values. However, complex expressions involving indexing or operators cannot be evaluated using this function

```python
#we need to convert this string of list to List
import ast
# ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# creating a helper function
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast']
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetchDirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            L.append(i['name'])
            
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetchDirector)
```

### we need to concate the list to form a large list ; convert that into a string (paragraph) which will be our tags

### we need to apply transformation
### 'Sam Worthington'     vs    'SamWorthington' 
#### If we donot remove the space than two tags will be created one for Sam and other for Worthington which may lead to confusion make it difficult for the model ; model can recommend other person named Sam instead of SamWorthington because two tags were created Similarly we need to remove spaces from all other columns

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#  Now we need to create tags column concatenating all columns other than id , title
movies['tags'] = movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

### We need to calculate consine distance to check for similarity(not using eulidean distance as not reliable for higher dimensions)
### Also distance is inversly propportional to similarity
### here we are using cosine_similarity method

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
sorted(list(enumerate(similarity[0])),reverse=True , key = lambda x:x[1])
```

#### In above sorting will change the index , so it will disturb the similarity matrix so to keep the index position will use enumerate function

##### similarity matrices is an array of arrays. Here row 0 tells distance of 1st movie with each other movies and son on for other rows

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True , key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

### The above recommennd is the main function
